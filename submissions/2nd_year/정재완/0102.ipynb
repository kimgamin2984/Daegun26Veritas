이번 차시에서는 똑같은 작업을 반복하는 **반복문**과, 코드를 재사용 가능한 단위로 묶는 **함수**를 배워요. 
설명은 주석을 참고하고, 각 셀의 [실습] 문제를 해결해 보세요.

```python
# [1] 반복문 (for): 지정된 횟수나 범위만큼 코드를 반복 실행함.
# range(시작, 끝) -> 시작부터 '끝-1'까지 숫자를 생성함.

print("--- 1부터 5까지 반복 ---")
for i in range(1, 6):
    print(i,"번째 실행 중...")

# 리스트 안의 요소들을 하나씩 꺼내올 수도 있음.
clubs = ["Veritas", "AI", "Python"]
for club in clubs:
    print("우리 동아리 이름:",club)
```
