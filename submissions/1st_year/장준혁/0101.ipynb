## 1단계: 변수(Variable)가 뭔가요?

컴퓨터에게 어떤 값을 기억시키고 싶을 때, 우리는 **'이름표가 붙은 그릇'** 을 사용합니다. 
이 그릇을 **'변수'** 라고 불러요.

### 📥 값을 넣는 법 (대입)
`a = 10` 이라는 코드는 수학의 "a와 10이 같다"는 뜻이 아니에요!
**"10이라는 데이터를 `a`라는 변수에 집어넣어라!"** 라는 명령입니다.

* **왼쪽(`a`)**: 그릇의 이름 (내 맘대로 지을 수 있음! 단, 숫자로 시작하거나 일부 정해진 단어는 안 됨)
* **오른쪽(`10`)**: 그릇에 담을 내용물
* **가운데(`=`)**: "집어넣어!" 라는 화살표 역할

```python
# 1. 'apple'이라는 그릇에 숫자 5를 담아봅시다.
apple = 5

# 2. 'banana'라는 그릇에 숫자 10을 담아봅시다.
banana = 10

# 3. 이제 그릇의 이름을 부르면 내용물이 나옵니다!
# 참고로 print()는 괄호 안의 내용물을 출력하라는 뜻이에요.
print(apple)
print(banana)
print(apple + banana) # 그릇 안의 내용물끼리 더할 수도 있어요.
```

### 1. 자료형(Data Type): 데이터의 세 가지 얼굴 🎭
가장 자주 쓰는 세 가지를 꼭 기억하세요!

1. **정수 (int)**: 딱 떨어지는 숫자 (예: 1, 100, -5)

## 2. 파이썬의 계산기 기능 (연산자)
파이썬은 아주 강력한 계산기입니다. 숫자형 데이터(int, float)를 가지고 다양한 산술 연산을 할 수 있어요.

### ➕ 산술 연산자
* `+`, `-`, `*`, `/` : 더하기, 빼기, 곱하기, 나누기
* `//` : **몫** (나눗셈 후 소수점 버림)
* `%` : **나머지** (홀짝 판별이나 배수 찾기에 필수!)
* `**` : **거듭제곱** (예: 2의 10제곱)

### ⚖️ 비교 연산자 (결과는 항상 True 아니면 False!)
* `==` : 같다 (수학의 =와 다름에 주의!)
* `!=` : 다르다
* `>`, `<`, `>=`, `<=` : 크기 비교

```python
# [실습] 직접 실행해서 결과를 예측해보세요.
a = 10
b = 3

print(f"10 나누기 3의 몫: {a // b}")
print(f"10 나누기 3의 나머지: {a % b}")
print(f"2의 10제곱: {2 ** 10}")

# 비교 연산 결과 확인
print(f"a와 b는 같은가? {a == b}")
print(f"a는 5보다 큰가? {a > 5}")
```

## 2. 리스트와 인덱싱, 슬라이싱
변수 하나에 한 개의 값만 집어넣는 건 너무 비효율적이죠.
변수 하나로 여러 개의 값을 다루는 법을 알아봅시다!

### 리스트
여러 개의 데이터를 순서대로 나란히 담아두는 바구니입니다. 파이썬에서 가장 많이 쓰이는 데이터 뭉치예요.  
대괄호 `[ ]` 안에 쉼표 `,`로 구분해요.

```python
fruits = ["사과", "배", "감", "복숭아", "살구"]
print(f"과일 바구니: {fruits}")
```

---
### 인덱싱(Indexing)
리스트에서 **몇 번째 데이터를 가져올지 번호로 콕 집는 것**입니다. 리스트 뒤에 띄어쓰기 없이 대괄호 `[ ]` 안에 원하는 수를 넣으면 돼요.
- **주의:** 파이썬은 숫자를 **0**부터 셉니다! (첫 번째 = 0번)
- **팁:** 음의 정수는 뒤에서부터 몇 번째인지를 뜻해요!

```python
# 0번(첫 번째) 과제 가져오기
print(f"0번 과일: {fruits[0]}")

# -1번(마지막) 과일 가져오기
print(f"마지막 과일: {fruits[-1]}")
```

---
### 3. 슬라이싱(Slicing)
리스트의 일부분을 **범위로 싹둑 잘라오는 것**입니다.
- **모양:** `리스트[시작 : 끝+1]`
- **주의:** `[1:3]`이면 1번부터 **200번**까지만 가져옵니다. (끝 번호는 포함 X)
- **팁:** 숫자를 생략하면 처음이나 끝으로 자동으로 지정해줘요!  
예를 들면 `리스트[:4]` 이건 리스트의 처음부터 4번 값까지, `리스트[1:]` 이건 리스트의 1번 값부터 끝까지를 지정하는 거죠,

```python
# 1번부터 3번까지(배, 감, 복숭아) 잘라오기
print(f"슬라이싱 결과: {fruits[1:4]}")

# 처음부터 2번까지 잘라오기
print(f"앞부분만: {fruits[:3]}")
```

## 4. 딕셔너리(Dictionary)
단어 그대로 **'사전'** 같은 구조입니다. 
- **모양:** `{키 : 값}` (중괄호 사용)
- **특징:** 번호(0, 1...)가 아니라 내가 정한 **이름표**로 데이터를 찾습니다.

### 5. 키(Key)와 밸류(Value)
- **키(Key):** 데이터를 찾기 위한 **이름표**
- **밸류(Value):** 이름표 뒤에 숨겨진 **실제 값**

```python
# 딕셔너리 만들기
me = {
    "이름": "김가민", 
    "나이": 16, 
    "소속": "대건고",
    "특기": ["파이썬", "데이터분석"] # 밸류에 리스트를 넣을 수도 있어요!
}
print(me)
```

### 항목 추가하기
딕셔너리에 원하는 키와 밸류을 추가하는 건 간단해요! 그냥 변수 대입하듯이 쓰면 됩니다.  
`dict[Key] = dict[Value]`  
(사실 원래 있는 키, 밸류 정보 수정도 위 코드로 돼요..)

```python
# 딕셔너리에 새로운 정보 추가하기
me["MBTI"] = "ENTP"
print(me)
```

---
### 6. 딕셔너리의 인덱싱
리스트는 번호`[n]`로 찾았지만, 딕셔너리는 **키(Key)** 이름으로 찾습니다.

```python
# '이름'이라는 키로 밸류 가져오기
print(f"나의 이름은? {me['이름']}")
print(f"나의 MBTI는? {me['MBTI']}")
```
