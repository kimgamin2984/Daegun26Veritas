### 1. 자료형(Data Type): 데이터의 세 가지 얼굴 🎭
가장 자주 쓰는 세 가지를 꼭 기억하세요!

1. **정수 (int)**: 딱 떨어지는 숫자 (예: 1, 100, -5)
2. **실수 (float)**: 소수점이 있는 숫자 (예: 3.14, 0.5)
3. **문자열 (str)**: 글자 덩어리. 반드시 **큰따옴표(" ")** 나 **작은따옴표(' ')** 로 감싸야 해요! (예: "베리타스", 'Python')

```python
# 자료형 확인하기: type() 이라는 마법 도구를 써봅시다.

a = 10
b = 3.14
c = "파이썬"

print(type(a))
print(type(b))
print(type(c))

# "10"은 숫자일까요, 문자일까요?
print(type("10"))
```

## 2. 파이썬의 계산기 기능 (연산자)
파이썬은 아주 강력한 계산기입니다. 숫자형 데이터(int, float)를 가지고 다양한 산술 연산을 할 수 있어요.

### ➕ 산술 연산자
* `+`, `-`, `*`, `/` : 더하기, 빼기, 곱하기, 나누기
* `//` : **몫** (나눗셈 후 소수점 버림)
* `%` : **나머지** (홀짝 판별이나 배수 찾기에 필수!)
* `**` : **거듭제곱** (예: 2의 10제곱)

### ⚖️ 비교 연산자 (결과는 항상 True 아니면 False!)
* `==` : 같다 (수학의 =와 다름에 주의!)
* `!=` : 다르다
* `>`, `<`, `>=`, `<=` : 크기 비교

```python
# [실습] 직접 실행해서 결과를 예측해보세요.
a = 10
b = 3

print(f"10 나누기 3의 몫: {a // b}")
print(f"10 나누기 3의 나머지: {a % b}")
print(f"2의 10제곱: {2 ** 10}")

# 비교 연산 결과 확인
print(f"a와 b는 같은가? {a == b}")
print(f"a는 5보다 큰가? {a > 5}")
```
